**How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?)**

```python
sal['Year'].unique()
select_2013 = sal['Year'] == 2013
sum(sal[select_2013]['JobTitle'].value_counts() == 1)
```

**How many people have the word Chief in their job title? (This is pretty tricky)**

```python
def check_chief(in_jobtitle): 
    if 'chief' in in_jobtitle.lower():
        return True
    else: 
        return False
```

```python
sum(sal['JobTitle'].apply(check_chief))
```

```python
sal['JobTitle'].str.lower().str.contains('chief')
```

## Data Visualization

**Implement by using seaborn's countplot with x='Year'**

```python
## plot the bar chart 
## illustrates the number of sample from 2011 to 2014 

## horizontal count axis
## vertical year axis

sns.countplot(x='Year', data=sal)
sns.countplot(y='Year', data=sal)
```

### Example
**Visualize top 5 Jobs with BasePay, Benefits and OvertimePay**

```python
## according to 'JobTitle'
## value_counts() -> Return a Series containing counts of unique values
## take top 5 

top_5_jobs = sal['JobTitle'].value_counts().head(5)
```

```python
## isin : return Boolean type
## Whether each element in the DataFrame is contained in values

select_top_5 = sal['JobTitle'].isin(top_5_jobs.index)
```

```python
## after checking
## assign top 5 from sal to sal2
sal2 = sal[select_top_5]

## groupby
## DataFrame or Series using a mapper or by a Series of columns
## mean()
sal3 = sal2.groupby('JobTitle').mean()
plot_df = sal3[['BasePay', 'Benefits', 'OvertimePay']]
```

```python
## plot()
## df.plot(kind='line') is equivalent to df.plot.line()
plot_df.plot(kind='bar')
```
